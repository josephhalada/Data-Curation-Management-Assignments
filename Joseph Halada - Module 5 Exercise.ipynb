```python
import pandas as pd
from pandas import DataFrame
```

```python
# Creation of dataframe, df that stores the assignment data.
data_dict = {"IMDB": [8.1, 7.4, 7.6, 7.9, 8.1], "Rotten Tomatoes": [96, 87, 90, 91, 90], "Metacritic": [75, 75, 76, 90, 74]}
df = DataFrame(data_dict, index = ["CODA", "Belfast", "King Richard", "Licorice Pizza", "Dune"])
df
```

```python
# Rename the index and columns of df, as indicated on the assignment.
df.index.name = "movies"
df.columns.name = "scores"
df
```

```python
# Using df to display metacritic scores of CODA and Dune.
df.loc[["CODA", "Dune"], ["Metacritic"]]
```

```python
# Sort df by movie names.
df.sort_index(axis = "index")
```

```python
# Sort df by score names.
df.sort_index(axis = "columns")
```

```python
# Sort df by IMDB scores in descending order.
df.sort_values(by = "IMDB", ascending=False)
```

```python
# Sort df first by Metacritic, next by Rotten Tomatoes scores.
df.sort_values(by = ["Metacritic", "Rotten Tomatoes"])
```

```python
# Update df to change IMDB score of Licorice Pizza to 8.5
df.loc["Licorice Pizza", "IMDB"] = 8.5
df
```

```python
# Drop King Richard from df
df.drop("King Richard", inplace=True)
df
```
