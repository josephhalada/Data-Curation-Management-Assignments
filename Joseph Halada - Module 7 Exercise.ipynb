```python
# import statements
import numpy as np
import pandas as pd
```

```python
# 1. Load indicated columns of csv file into a dataframe, located at https://raw.githubusercontent.com/ryurko/nflscrapR-data/master/play_by_play_data/post_season/post_pbp_2009.csv
df = pd.read_csv("https://raw.githubusercontent.com/ryurko/nflscrapR-data/master/play_by_play_data/post_season/post_pbp_2009.csv", usecols=["play_id", "game_id", "home_team", "away_team", "game_date", "play_type", "replay_or_challenge", "replay_or_challenge_result"])
df
```

```python
# 2. Count number of missing entries in this dataframe.
df.isna().sum().sum()
```

There are 2157 missing entries in this dataframe.

```python
# 3. Percentage of entries missing.
percent_missing = df.isnull().sum() * 100 / len(df)
percent_missing
```

3.33% of entries are missing from play_type, and 99.29% of entries are missing from replay_or_challenge_result

```python
# 4. If you removed all the rows containing a missing value, how many rows would remain?
len(df.dropna())
```

15 rows would remain if one removed all the rows containing a missing value.

```python
# 5. If you removed all the columns containing a missing value, how many columns would remain?
len(df.dropna(axis=1).columns)
```

6 columns would remain if you dropped all columns containing a missing value.

```python
# 6. Repeat Step 1, if your dataframe has changed. Replace all NaN values in the play_type column with ‘pass’.
df = pd.read_csv("https://raw.githubusercontent.com/ryurko/nflscrapR-data/master/play_by_play_data/post_season/post_pbp_2009.csv", usecols=["play_id", "game_id", "home_team", "away_team", "game_date", "play_type", "replay_or_challenge", "replay_or_challenge_result"])
df.fillna({"play_type": "pass"}, inplace=True)
df
```

```python
# 7. Make all the values in play_type column UPPER CASE.
df["play_type"] = df["play_type"].str.upper()
df
```

```python
# 8. How many distinct play types exist in your data?
len(df["play_type"].unique())
```

9 unique play types exist in my data.
