```python
# import statements
import pandas as pd
import numpy as np

# assignment setup
df1 = pd.DataFrame({
    'id': range(7),
    'Title': ['Toy Story', 'A Bug\'s Life', 'Toy Story 2', 'Monsters, Inc.', 'Finding Nemo', 'Inside Out', 'The Good Dinosaur'],
    'Director': ['John Lasseter', 'John Lasseter', 'John Lasseter', 'Pete Docter', 'Andrew Stanton', 'Pete Docter', 'Peter Sohn'],
    'Year': [1995, 1998, 1999, 2001, 2003, 2015, 2015],
    'Length_minutes': [81, 95, 93, 92, 107, 95, 97]
    })

df2 = pd.DataFrame({
    'movie_id': [0, 0, 1, 2, 4, 4, 4, 5, 6, 8],
    'critic_rating': [8.3, 7.2, 7.9, 8.1, 8.2, 8.9, 9.1, 8.3, 8.9, 9.0]
})

df3 = pd.DataFrame({
    'movie_id': range(4),
    'Domestic_sales_millions': [191, 162, 245, 289],
    'International_sales_millions': [170, 200, 239, 272]
})
```

```python
# Create a table that shows for each movie, its movie title, year and international sales. 
# For the movies where international sales is not available, show null instead.
df4 = pd.concat([df1, df3], axis=1)
df4.drop({"id", "Director", "Length_minutes", "movie_id", "Domestic_sales_millions"}, axis=1, inplace=True)
df4.replace({np.nan: "null"})
```

```python
# Create a table showing for each movie, its title, its critic ratings (show null if not available) and its domestic sales. 
# Don't include movies for which domestic sales is not available
df5 = pd.concat([df1, df2, df3], axis = 1)
df5.drop({"id", "Director", "Year", "Length_minutes", "movie_id", "International_sales_millions"}, axis=1, inplace=True)
df5.dropna(inplace=True)
df5
```

```python
# Create a new dataframe from df1 that sets 'Director' and 'Year' to be the joint index.
df6 = df1
df6.set_index(["Director", "Year"], inplace=True)
df6
```

```python
# Using the dataframe created in the last step, calculate for each director, the average length (in minutes) of their movies.
df6.groupby("Director").mean("Length_minutes")
```

```python
# Using the dataframe created in Step 3, print only the movies directed by John Lasseter or Pete Docter.
# Show only the columns for movie id and its length
df6.loc[(["John Lasseter", "Pete Docter"]), ["id", "Length_minutes"]]
```

```python
# Using the dataframe created in Step 3, print only the movies directed in the years 1995 or 2015. 
# Show only the columns for movie id and its title
df7 = df6.query("Year == 1995 or Year == 2015")
df7.drop({"Length_minutes"}, axis=1, inplace=True)
df7
```
