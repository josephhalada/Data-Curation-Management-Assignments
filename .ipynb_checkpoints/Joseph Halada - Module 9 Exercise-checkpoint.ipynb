```python
# import statement
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# histogram of sepal length as indicated in assignment.
df = pd.read_csv("iris.data")
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.hist(df["sepal length"],label="Sepal length", color="#2ab0fc", edgecolor="magenta", linewidth=3.5)
ax.set_title("Sepal length")
ax.legend(loc="upper left")
ax.set_xlabel("Bins")
ax.set_ylabel("Centimeters")
```

```python
# There are three species of flowers in the dataset. 
#The sepal length-width combination of one of the species is quite distinct from that of the other two species.
# Make an appropriate visualization plot to find out the distinct species. 
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.scatter(df["species"], (df["sepal length"] / df["sepal width"]))
ax.set_title("Species and sepal length")
ax.set_xlabel("Species")
ax.set_ylabel("Sepal Length:Width Ratio")
```

```python
# Make a plot that shows you the distribution of petal length for each of the three species.
# It should show median, 25th and 75th percentile values and outliers.
sns.catplot(
    x ="petal length",
    y ="species",
    kind ="box",
    data=df)
```
