```python
# import statements
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

# read csv and create dataframe
titanic_df = pd.read_csv("titanic.csv")
titanic_df
```

```python
# Write a Pandas program to split the titanic dataframe by passenger class ('class') 
# and get mean value, median value, min value, and max value of (1) age, (2) fare for each passenger class.
# Also, calculate for each passenger class, how many passengers' age and fare are known 
# (i.e., have non-null entries) in the same output dataframe.

print("Average age and fare of each passenger class:")
display(titanic_df.groupby("class")[["age", "fare"]].mean())
print("Median age and fare of each passenger class:")
display(titanic_df.groupby("class")[["age", "fare"]].median())
print("Minimum age and fare of each passenger class:")
display(titanic_df.groupby("class")[["age", "fare"]].min())
print("Maximum age and fare of each passenger class:")
display(titanic_df.groupby("class")[["age", "fare"]].max())
```

```python
# What is the probability of survival for each passenger class?
titanic_df.groupby("class")["survived"].value_counts(1, sort=False)
```

First class had a 62.9% survival rate.
Second Class had a 47.2% survival rate.
Third Class had a 24.2% survival rate.

```python
# What is the probability of survival for men, women, and children (Hint: Use column 'who')?
titanic_df.groupby("who")["survived"].value_counts(1, sort=False)
```

Children had a 59% survival rate.
Men had a 16.3% survival rate.
Women had a 75.6% survival rate.

```python
# For each combination of town of embarkment and deck, calculate
# (1) the number of people who belonged to the combination (if any) and (2) the number of people who survived.
display(titanic_df.groupby("deck")["embark_town"].value_counts())
```

```python
# Draw a category plot (bar plot) with passenger class on x-axis, 'survived' on y-axis, and gender as hue. 
# Explain in your own words what insights this plot gives you about the probability of survival.
titanic_barplot = sns.barplot(x="class", y="survived", hue="sex", data=titanic_df, order=["First", "Second", "Third"])
```

One can examine that regardless of gender, the rate of survival drops sharply the lower your class was. In addition, not even first class male passengers had higher survival rates than third class female passengers. Due to the circumstances, it can likely be assumed that a large chunk of these male passengers are children, due to the preference given to women and children.
