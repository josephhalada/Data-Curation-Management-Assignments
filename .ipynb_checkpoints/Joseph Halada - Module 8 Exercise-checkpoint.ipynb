```python
import pandas as pd
import numpy as np
import networkx as nx
```

```python
# Read it with python. Set the first column as the index when reading the file.
df = pd.read_csv("social_indicator.txt", sep="	", index_col="Index")
df
```

```python
# Sort the row index into ascending order.
df.sort_index(inplace=True)
df
```

```python
# Make a column â€˜westâ€™ with True and False values indicating whether the region is west. Then delete the original region column.
df["west"] = df["region"]
df["west"].replace({"west": True}, inplace=True)
df["west"].replace({"other": False}, inplace=True)
df.drop("region", axis=1, inplace=True)
df
```

```python
# Suppose we are not interested in studying fertility rate (â€˜totalfertilityrateâ€™) and contraception (â€˜contraceptionâ€™).
# Drop these two columns.
df.drop("totalfertilityrate", axis=1, inplace=True)
df.drop("contraception", axis=1, inplace=True)
df
```

```python
# Use the column â€˜geographic_locationâ€™ to create a new column called â€˜continentâ€™. 
# Merge the values of the â€˜geographic column' appropriately to obtain 6 distinct values for the â€˜continentâ€™ column 
# â€“ Asia, Africa, North America, South America, Europe and Oceania. Drop the column â€˜geographic_locationâ€™.
df["continent"] = df["geographic_location"]
df["continent"] = df["continent"].replace({"Northern": ""}, regex=True)
df["continent"] = df["continent"].replace({"North-eastern": ""}, regex=True)
df["continent"] = df["continent"].replace({"Eastern": ""}, regex=True)
df["continent"] = df["continent"].replace({"South-eastern": ""}, regex=True)
df["continent"] = df["continent"].replace({"Southern": ""}, regex=True)
df["continent"] = df["continent"].replace({"South-western": ""}, regex=True)
df["continent"] = df["continent"].replace({"Western": ""}, regex=True)
df["continent"] = df["continent"].replace({"North-western": ""}, regex=True)
df["continent"] = df["continent"].replace({"Middle": ""}, regex=True)
df["continent"] = df["continent"].replace({"Central America": ""}, regex=True)
df["continent"] = df["continent"].replace({"Central": ""}, regex=True)
df.drop("geographic_location", axis=1, inplace=True)
df
```

```python
# Create a complete graph with European country names as nodes.
def all_pairs_digraph(nodes):
    for i, u in enumerate(nodes):
        for j, v in enumerate(nodes):
            if i < j:
                yield u, v
                yield v, u

def make_complete_directed_graph(n):
    G = nx.DiGraph()
    nodes = n
    G.add_nodes_from(nodes)
    G.add_edges_from(all_pairs_digraph(nodes))
    return G

complete_graph = make_complete_directed_graph(df["country"])

nx.draw_circular(complete_graph, 
                 node_color='C2', 
                 node_size=1000, 
                 with_labels=True)
```
