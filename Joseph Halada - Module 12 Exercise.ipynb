```python
# import statements
import pandas as pd
import numpy as np

# create spotify_df using the spotify_data.csv file
spotify_df = pd.read_csv("spotify_data.csv")
spotify_df
```

```python
# Which three genres have the highest danceability on average?
display(spotify_df.groupby("genres")[["danceability"]].mean())
```

rap has the highest danceability, followed by hip hop, which is then followed by hoerspiel

```python
# Use 'track_popularity' to create a new column 'track_popularity_level' by binning track_popularity into 3 bins. 
# Label the bins 'low', 'medium' and 'high' (Hint: Use pd.cut).
spotify_df["track_popularity_level"] = pd.cut(spotify_df["track_popularity"], 3, labels=["low", "medium", "high"])
spotify_df
```

```python
# Create a pivot table to show the average danceability of tracks grouped by genre and track 
# popularity level on the rows and separated by energy levels in the columns. 
# Show the margin averages. What does this pivot table tell you about the relationship between 
# popularity and danceability for a) country tracks and b) electronic tracks?
spotify_pivot = spotify_df.pivot_table("danceability", index="track_popularity_level", columns="genres", aggfunc="mean", margins=True)
display(spotify_pivot)
```

This shows me that less popular country tracks are more danceable than more popular country tracks, while the opposite applies for electronic tracks. More popular electronic tracks are more danceable than less popular electronic tracks.

```python
# What percentage of all the unique tracks are contributed by the top 3 artists of each genre, 
# where the top artists are based on artist_popularity, and the unique tracks are based on unique values of track_name?
top_artists_genre = spotify_df.groupby(["genres", "artist_name"])[["artist_popularity"]].max()
top_artists_track_count = spotify_df.groupby(["genres", "artist_name"])[["track_name"]].nunique()
display(top_artists_genre)
display(top_artists_track_count)
```
